```
data.dropna(inplace=True)
```

```
from sklearn.model_selection import train_test_split
```

```
#the .iloc[]function for primarily integer position based 
```

```
X = data.iloc[:,1:260]
Y = data.iloc[:,38]
```

```
#Spliting the inputs and the outputs into  the train_test_split() function
```

```
x_train , x_t , y_train, y_t = train_test_split(X, Y, test_size=0.2, random_state=42) 
```

```
#importation of the Support vector machines and the classification model  
```

```
from sklearn import svm 
```

```
#the C-Support Vector Classification
```

```
clf = svm.SVC()
#Fit the SVM model according to the given training data (Trainig)
clf.fit(x_train, y_train)
```

```
#Performing the classification on samples in X_train.
```

```
pred = clf.predict(x_t)
```

```
from sklearn import metrics
```

```
#the F1 score as evaluation metric
```

```
metrics.f1_score(y_t, pred)
```
